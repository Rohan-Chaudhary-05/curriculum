**Finding Outliers & Extreme Values**

```python
# Reloading our data
data = pd.read_csv('/content/Prepared_Mall_Customers.csv')
data.describe().transpose()
# We used transpose to make the columns rows and the rows columns to twist the table
# UNEXPECTED OUTPUT
```

```python
# Drop unnecessary variables and rename your dataset
df = data.drop(columns=(['ID', 'NewColumn']))
df.describe()
```

```python
# Investigating histogram for potential outliers
Salary_fig = px.histogram(df, x='Salary')
Salary_fig.show()
```

```python
# Use of box plot allows us to observe univariate outliers
Age_fig = px.box(df, x='Age')
Age_fig.show()
```

```python
# Attempting to do the same but for Salary
Salary_fig = px.box(df, x='Salary')
Salary_fig.show()
```

```python
# Now we plot the data on a scatter to view inconsistent datapoints
# Can cross check with box plots to ensure datapoints are within range
# The 2 points around age 30 and salary 137k seem inconsistent but do meet box plot values hence may not be outliers
Age_Salary_Scatter_fig = px.scatter(x=df['Age'], y=df['Salary'])
Age_Salary_Scatter_fig.show()
```

```python
# Create function which uses formula to find outliers easier
def find_outliers_IQR(df):
  q1=df.quantile(0.25)
  q3=df.quantile(0.75)
  IQR=q3-q1
  outliers = df[((df<(q1-1.5*IQR))|(df>(q3+1.5*IQR)))]
  return outliers
```

```python
# Run function to find 'Salary' outliers
outliers = find_outliers_IQR(df['Salary'])
print("number of outliers: "+ str(len(outliers)))
outliers
```

```python
# We now drop the outliers
df.drop(df.index[[199,198]], inplace=True)

# Re-run function to see ensure outliers have been removed
outliers = find_outliers_IQR(df['Salary'])
print("number of outliers: "+ str(len(outliers)))
outliers
```

```python
# Now we have only the relevant columns and a new 'Salary' max value
df.describe().transpose()
```
